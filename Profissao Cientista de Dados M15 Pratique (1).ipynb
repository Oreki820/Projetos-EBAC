# 3) Identifique e trate as colunas que contém outliers.
Caso opte por mante-los ou altera-los justifique sua escolha.


```python
#Seu código aqui
def detectar_outliers_iqr(df, coluna):
    Q1 = df[coluna].quantile(0.25)
    Q3 = df[coluna].quantile(0.75)
    IQR = Q3 - Q1
    limite_inferior = Q1 - 1.5 * IQR
    limite_superior = Q3 + 1.5 * IQR
    outliers = df[(df[coluna] < limite_inferior) | (df[coluna] > limite_superior)]
    return outliers

# Exemplo para total_pago
outliers_total_pago = detectar_outliers_iqr(df, 'total_pago')
print(f'Outliers em total_pago: {len(outliers_total_pago)} registros')
```

Após aplicar a técnica do Intervalo Interquartílico (IQR) para identificar outliers nas variáveis numéricas (total_pago, pagamento_mensal, tempo_como_cliente), nenhum registro foi identificado como outlier extremo.

Por isso, optei por manter os dados originais, pois os valores altos refletem comportamentos reais dos clientes, como longa fidelidade ou assinatura de planos mais caros. Alterá-los ou removê-los poderia comprometer a integridade da análise preditiva, especialmente no contexto de churn.
